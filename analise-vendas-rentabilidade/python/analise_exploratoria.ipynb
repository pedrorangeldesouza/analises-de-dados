```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vendas.csv')
```

```python
df['Receita'] = df['Quantidade'] * df['Preco']
df['Custo_Total'] = df['Quantidade'] * df['Custo']
df['Lucro'] = df['Receita'] - df['Custo_Total']
df['Margem'] = df['Lucro'] / df['Receita']
```

```python
lucro_categoria = (
    df.groupby('Categoria')['Lucro']
        .sum()
        .sort_values(ascending=False)
)
```

```python
plt.figure()
lucro_categoria.plot(kind='bar')
plt.title('Lucro Total por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Lucro')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
